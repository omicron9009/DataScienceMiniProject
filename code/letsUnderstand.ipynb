```python
import pandas as pd

# ---- MISSING VALUE ANALYSIS ---- #
def missing_value_report(df: pd.DataFrame) -> pd.DataFrame:
    """
    Generate a clean, well-organized summary of missing values in a DataFrame.
    
    Returns:
        pd.DataFrame: Columns, Missing Count, Percentage Missing, and Data Type.
    """
    total = df.isnull().sum()
    percent = (total / len(df)) * 100
    dtype = df.dtypes

    missing_df = (
        pd.concat([total, percent, dtype], axis=1)
        .reset_index()
        .rename(columns={
            'index': 'Column',
            0: 'Missing_Values',
            1: 'Percent_Missing',
            2: 'Data_Type'
        })
        .query("Missing_Values > 0")
        .sort_values(by='Percent_Missing', ascending=False)
        .reset_index(drop=True)
    )

    return missing_df

# ---- USAGE ---- #
missing_summary = missing_value_report(data)
print("üîç Missing Value Summary:\n")
print(missing_summary.to_string(index=False))
```
